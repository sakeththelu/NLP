## Load Text

### Subtask:
Define a short paragraph of text as a string within the notebook for processing.


```python
text_paragraph = "Natural language processing (NLP) is a field of artificial intelligence that focuses on the interaction between computers and human language. It involves various tasks such as text classification, sentiment analysis, and machine translation."
print(f"Defined text paragraph: {text_paragraph[:50]}...")
```

## Convert to Lowercase

### Subtask:
Convert the loaded paragraph of text to all lowercase characters using Python's built-in string methods.


```python
lowercase_text = text_paragraph.lower()
print(f"Converted to lowercase (first 50 chars): {lowercase_text[:50]}...")
```

## Count Words

### Subtask:
Count the total number of words in the preprocessed (lowercase) text.


```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize

# Tokenize the lowercase text into words
words = word_tokenize(lowercase_text)

# Count the total number of words
word_count = len(words)

print(f"Total word count: {word_count}")
```

```python
import nltk
nltk.download('punkt')

# Count the number of words in the text
words = nltk.word_tokenize(lowercase_text)
word_count = len(words)
print(f"Number of words in the text: {word_count}")
```
