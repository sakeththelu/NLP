```python
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
text = """Loving the new AI  features

 #AI #MachineLearning"""
doc = nlp(text)
nouns = []
verbs = []
for token in doc:
    if token.pos_ in ["NOUN", "PROPN"]:
        nouns.append(token.text)
    elif token.pos_ == "VERB":
        verbs.append(token.text)
noun_freq = Counter(nouns)
verb_freq = Counter(verbs)
print("Noun Frequency:", noun_freq)
print("Verb Frequency:", verb_freq)
```

```python
import nltk
import spacy
from collections import Counter

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

text = """I am a software professional working as a backend developer, with a strong interest in building practical and reliable applications. I enjoy learning new technologies and strengthening my core concepts in areas like web development, security, and system fundamentals. Along with my job, I actively prepare for academic exams by focusing on clear and simplified understanding of topics. I consider myself curious, hardworking, and goal-oriented, always looking to improve my skills and grow as a developer."""

print("Essay Text:\n", text)

tokens = nltk.word_tokenize(text)
print("\nTokens:\n", tokens)


nltk_pos_tags = nltk.pos_tag(tokens)
print("\nNLTK POS Tags:\n", nltk_pos_tags)

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

spacy_pos_tags = [(token.text, token.pos_) for token in doc]
print("\nspaCy POS Tags:\n", spacy_pos_tags)


nouns = [token.text.lower() for token in doc if token.pos_ == "NOUN"]
verbs = [token.text.lower() for token in doc if token.pos_ == "VERB"]

noun_freq = Counter(nouns)
verb_freq = Counter(verbs)

print("\nNoun Frequency (Concepts):")
for noun, count in noun_freq.most_common():
    print(f"{noun}: {count}")

print("\nVerb Frequency (Arguments):")
for verb, count in verb_freq.most_common():
    print(f"{verb}: {count}")


print("\nComparison:")
print("NLTK uses detailed POS tags (NN, VB, JJ, etc.)")
print("spaCy uses simplified universal tags (NOUN, VERB, ADJ)")
print("spaCy is more suitable for academic vocabulary analysis")

print("\nDiscussion:")
print("Nouns represent key academic concepts such as developer, technologies, and skills.")
print("Verbs indicate actions and arguments such as learning, working, and improving.")
print("POS tagging helps analyze structure, focus, and clarity of academic writing.")

print("\nConclusion:")
print("POS tagging using NLTK and spaCy effectively supports academic text analysis.")
```
